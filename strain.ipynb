```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d as Axes3D
```

```python
tensor = [[1, 0, 0], 
          [0, 1, 0], 
          [0, 0, 1]]
```

```python
points = np.array([[0, 0, 0], [1, 0, 0], [0, 1, 0], [1, 1, 0]])
points
```

```python
fig: plt.Figure = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(points[:,0], points[:, 1], points[:, 2])

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.ion()

plt.show()
```

```python
def get_modified_coordinate(x, tensor_row, point_):
    new_x = x
    for j in range(3):
        new_x += tensor_row[j] * point_[j]
    return new_x

def get_modified_point(point_, tensor_):
    new_point = [0, 0, 0]
    for i in range(3):
        new_point[i] = get_modified_coordinate(point_[i], tensor_[i], point_)
    return new_point
```

```python
new_points = []

for point in points:
    new_points.append(get_modified_point(point, tensor))

new_points = np.array(new_points)
new_points
```

```python
fig: plt.Figure = plt.figure()
ax = fig.add_subplot(111, projection='3d')
#ax.scatter(points[:,0], points[:, 1], points[:, 2], c='b')
ax.scatter(new_points[:, 0], new_points[:, 1], new_points[:, 2], c='r')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.ion()

plt.show()
```
